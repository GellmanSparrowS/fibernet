# Getting Started with FiberNet

This tutorial introduces the basic concepts of FiberNet and shows you how to:

1. Generate fiber networks (random, ordered, chiral, woven)
2. Analyze network properties
3. Visualize networks
4. Run simple simulations

````python
import numpy as np
import matplotlib.pyplot as plt
from fibernet import gen
from fibernet.analysis import MorphologyAnalyzer
from fibernet.viz import visualize_3d_matplotlib
````

## 1. Generate a Random 2D Fiber Network

The simplest way to create a fiber network is using the random generator:

````python
# Generate a random 2D fiber network
net = gen.random_straight_2d(
    num_fibers=100,
    fiber_length=15.0,
    box_size=(50, 50),
    radius=0.1,
    seed=42
)

print(f'Fibers: {net.num_fibers}')
print(f'Crosslinks: {net.num_crosslinks}')
print(f'Dimension: {net.dimension}D')
````

## 2. Analyze Network Properties

FiberNet provides comprehensive analysis tools:

````python
# Create analyzer
analyzer = MorphologyAnalyzer(net)

# Compute key metrics
mean_length = analyzer.mean_fiber_length()
nematic_order = analyzer.nematic_order_parameter()
porosity = analyzer.porosity()
connectivity = analyzer.mean_connectivity()

print(f'Mean fiber length: {mean_length:.2f}')
print(f'Nematic order parameter: {nematic_order:.3f}')
print(f'Porosity: {porosity:.3f}')
print(f'Mean connectivity: {connectivity:.2f}')
````

## 3. Visualize the Network

Use matplotlib for static plots or pyvista for interactive 3D:

````python
# Static 3D plot
fig, ax = visualize_3d_matplotlib(net, show_crosslinks=True)
plt.title('Random 2D Fiber Network')
plt.show()
````

## 4. Different Network Types

FiberNet supports many types of fiber networks:

````python
# Ordered square lattice
lattice = gen.square_lattice_2d(spacing=5.0, grid_size=(10, 10))

# Honeycomb lattice
honeycomb = gen.honeycomb_lattice_2d(spacing=5.0, grid_size=(5, 5))

# Chiral network (helical fibers)
chiral = gen.chiral_network_2d(
    num_fibers=50,
    fiber_length=15.0,
    box_size=(50, 50),
    chirality=0.5,
    seed=42
)

# Woven structure
woven = gen.woven_2d(
    warp_count=10,
    weft_count=10,
    spacing=5.0,
    weave_pattern='plain'
)

print(f'Lattice: {lattice.num_fibers} fibers')
print(f'Honeycomb: {honeycomb.num_fibers} fibers')
print(f'Chiral: {chiral.num_fibers} fibers')
print(f'Woven: {woven.num_fibers} fibers')
````

## 5. Compare Different Network Types

Let's compare the nematic order parameter across different network types:

````python
networks = {
    'Random': net,
    'Lattice': lattice,
    'Honeycomb': honeycomb,
    'Chiral': chiral,
}

fig, ax = plt.subplots(1, 1, figsize=(8, 5))

names = []
orders = []

for name, network in networks.items():
    analyzer = MorphologyAnalyzer(network)
    order = analyzer.nematic_order_parameter()
    names.append(name)
    orders.append(order)

ax.bar(names, orders, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
ax.set_ylabel('Nematic Order Parameter')
ax.set_title('Fiber Alignment Comparison')
ax.set_ylim(0, 1)
plt.tight_layout()
plt.show()
````

## 6. 3D Networks

FiberNet also supports 3D fiber networks:

````python
# Generate a 3D random network
net_3d = gen.random_straight_3d(
    num_fibers=200,
    fiber_length=20.0,
    box_size=(50, 50, 50),
    seed=42
)

print(f'3D Network: {net_3d.num_fibers} fibers, {net_3d.num_crosslinks} crosslinks')

# Analyze
analyzer_3d = MorphologyAnalyzer(net_3d)
print(f'3D Porosity: {analyzer_3d.porosity():.3f}')
````

## 7. Saving and Loading Networks

Networks can be saved and loaded in various formats:

````python
from fibernet.io import save_json, load_json, to_vtk

# Save as JSON
save_json(net, 'my_network.json')

# Load from JSON
loaded_net = load_json('my_network.json')
print(f'Loaded network: {loaded_net.num_fibers} fibers')

# Export to VTK for ParaView
to_vtk(net, 'network.vtk')
print('Exported to VTK format')
````
